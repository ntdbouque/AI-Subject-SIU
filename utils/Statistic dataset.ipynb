```python
# import libraries
import os
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# thống kê kích thước:
dataset_path = "/content/drive/MyDrive/Dataset/School AI-Project/augumented_dataset"

image_sizes = []

for class_name in os.listdir(dataset_path):
  class_path = os.path.join(dataset_path, class_name)
  lst_img_paths = glob.glob(os.path.join(class_path, "*.png"))
  for img_path in lst_img_paths:
    image = cv2.imread(img_path)
    height, width,_ = image.shape
    image_sizes.append((height, width))
```

## 1. Compost

```python
# duy
def create_lst_size(dataset_path):
  lst_size = []
  lst_img_paths = glob.glob(os.path.join(dataset_path, "*.png"))
  for img_path in lst_img_paths:
      image = cv2.imread(img_path)
      height, width,_ = image.shape
      compost.append((height, width))

  return lst_size
```

```python
compost = create_lst_size("/content/drive/MyDrive/Dataset/School AI-Project/augumented_dataset/Compost")
```

```python
def statistics_height_width(lst_size):
  # Chuyển đổi list thành mảng NumPy để thuận tiện trong tính toán
  image_sizes_array = np.array(lst_size)

  # Tính trung bình, min, max và độ lệch chuẩn cho chiều dài
  mean_height = np.mean(image_sizes_array[:, 0])
  min_height = np.min(image_sizes_array[:, 0])
  max_height = np.max(image_sizes_array[:, 0])
  std_height = np.std(image_sizes_array[:, 0])

  # Tính trung bình, min, max và độ lệch chuẩn cho chiều rộng
  mean_width = np.mean(image_sizes_array[:, 1])
  min_width = np.min(image_sizes_array[:, 1])
  max_width = np.max(image_sizes_array[:, 1])
  std_width = np.std(image_sizes_array[:, 1])

  # Hiển thị kết quả
  print("Trung bình chiều dài:", mean_height)
  print("Chiều dài nhỏ nhất:", min_height)
  print("Chiều dài lớn nhất:", max_height)
  print("Độ lệch chuẩn chiều dài:", std_height)

  print("************")
  print("Trung bình chiều rộng:", mean_width)
  print("Chiều rộng nhỏ nhất:", min_width)
  print("Chiều rộng lớn nhất:", max_width)
  print("Độ lệch chuẩn chiều rộng:", std_width)
```

```python
statistics_height_width(compost)
```

```python
def count_img(lst_size):
  # Tính trung bình cho chiều dài và chiều rộng
  duy = np.array(lst_size)
  mean_height = np.mean(duy[:, 0])
  mean_width = np.mean(duy[:, 1])

  # Đếm số lượng ảnh có chiều dài nhỏ hơn trung bình và chiều rộng tương tự
  count_below_mean_height = np.sum(duy[:, 0] < mean_height)
  count_below_mean_width = np.sum(duy[:, 1] < mean_width)

  # Hiển thị kết quả
  print("Số lượng ảnh có chiều dài nhỏ hơn trung bình:", count_below_mean_height)
  print("Số lượng ảnh có chiều rộng nhỏ hơn trung bình:", count_below_mean_width)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Biểu đồ histogram cho chiều dài
axes[0, 0].hist(image_sizes_array[:, 0], bins=20, color='blue', alpha=0.7)
axes[0, 0].set_title('Histogram Chiều Dài')

# Biểu đồ histogram cho chiều rộng
axes[0, 1].hist(image_sizes_array[:, 1], bins=20, color='green', alpha=0.7)
axes[0, 1].set_title('Histogram Chiều Rộng')

# Biểu đồ thanh cho trung bình, min, max và độ lệch chuẩn chiều dài
axes[1, 0].bar(['Trung bình', 'Min', 'Max', 'Độ lệch chuẩn'], [mean_height, min_height, max_height, std_height], color='blue')
axes[1, 0].set_title('Thống kê Chiều Dài')

# Biểu đồ thanh cho trung bình, min, max và độ lệch chuẩn chiều rộng
axes[1, 1].bar(['Trung bình', 'Min', 'Max', 'Độ lệch chuẩn'], [mean_width, min_width, max_width, std_width], color='green')
axes[1, 1].set_title('Thống kê Chiều Rộng')

plt.tight_layout()
plt.show()
```

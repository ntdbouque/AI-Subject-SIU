```python
# Declare an augmentation pipeline
transform = A.Compose([
    #A.RandomCrop(width=256, height=256),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
])
```

```python
def augmentation(dataset_path, save_path):
    for class_name in os.listdir(dataset_path):
        class_name_path = os.path.join(dataset_path, class_name)
        lst_img_paths = glob.glob(os.path.join(class_name_path, "*.png"))

        # Tạo thư mục đích nếu nó chưa tồn tại
        des_path = os.path.join(save_path, class_name)
        os.makedirs(des_path, exist_ok=True)

        for img_path in lst_img_paths:
          for i in range(20):
              image = cv2.imread(img_path)
              image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

              transformed = transform(image=image)
              transformed_image = transformed["image"]

              basename = os.path.basename(img_path)
              save_img_path = os.path.join(des_path, basename)

              boolean = cv2.imwrite(save_img_path, cv2.cvtColor(transformed_image, cv2.COLOR_RGB2BGR))
              print(f"Đã ghi ảnh {basename}: {boolean}")
```

```python
dataset_path = "/content/drive/MyDrive/Dataset/School AI-Project/use"
save_path = "/content/drive/MyDrive/Dataset/School AI-Project/output"
augmentation(dataset_path, save_path)
```
